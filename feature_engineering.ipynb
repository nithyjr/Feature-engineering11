1. What is a parameter?

A parameter is a value learned by a machine learning model during training that helps make predictions.
Example: In linear regression, weights (coefficients) and bias are parameters.

2. What is correlation?

Correlation measures the relationship between two variables—how one variable changes when another changes.

3. What does negative correlation mean?

Negative correlation means:

When one variable increases, the other decreases
Example: As price increases, demand decreases.

4. Define Machine Learning. What are the main components?

Machine Learning (ML) is a field of AI where systems learn patterns from data and make predictions without being explicitly programmed.

Main components:

Data

Features

Model

Loss function

Optimizer

Evaluation metrics

5. How does loss value help determine model quality?

Loss value measures how far predictions are from actual values.

Lower loss = better model

High loss indicates poor learning.

6. What are continuous and categorical variables?

Continuous variables: Numeric values with range
Example: Height, Weight, Marks

Categorical variables: Categories or labels
Example: Gender, Course, City

7. How do we handle categorical variables? Common techniques

Label Encoding – Converts categories to numbers

One-Hot Encoding – Creates binary columns

Ordinal Encoding – For ordered categories

8. What is training and testing a dataset?

Training set: Used to train the model

Testing set: Used to evaluate model performance on unseen data

9. What is sklearn.preprocessing?

A module in scikit-learn used for:

Feature scaling

Encoding categorical data

Normalization

10. What is a Test set?

A test set is a portion of data kept aside to evaluate how well the trained model performs on new data.

15. What is causation? Correlation vs causation

Causation: One variable directly causes another

Correlation: Variables move together, but no cause

Example:

Ice cream sales and drowning cases increase together
➡ Correlated, but ice cream does NOT cause drowning

16. What is an Optimizer? Types with examples

An optimizer updates model parameters to minimize loss.

Types:

Gradient Descent

Stochastic Gradient Descent (SGD)

Adam

RMSprop

Example:
Adam optimizer is widely used in deep learning for fast convergence.

20. What is feature scaling? Why is it needed?

Feature scaling brings all features to the same range.

Helps:

Faster convergence

Better performance for distance-based models

21. How do we perform scaling in Python?
